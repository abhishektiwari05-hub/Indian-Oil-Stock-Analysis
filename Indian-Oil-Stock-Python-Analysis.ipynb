```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv(r"C:\Users\at434\OneDrive\Desktop\python\Indian Oil Data (2).csv")
df.head()
```

```python
# Highest Closing Price

df[df["Close"]==df["Close"].max()]
```

```python
# Lowest Closing Price

df[df["Close"]==df["Close"].min()]
```

```python
# Average Closing Price

df["Close"].mean()
```

```python
# Highest Trading Volume

df.sort_values("Volume",ascending=False).head(10)
```

```python
# Daily Return

df["Daily Return"]=(df["Close"]-df["Open"])/df["Open"]*100
```

```python
# Price Difference

df["Price Change"]=df["Close"]-df["Open"]
```

```python
# Top 10 Highest Gain Days

df.sort_values("Price Change",ascending=False).head(10)
```

```python
# Top 10 Highest Loss Days

df.sort_values("Price Change").head(10)
```

```python
# Monthly Average Close

df.groupby("Month")["Close"].mean()
```

```python
# Yearly Average Close

df.groupby("Year")["Close"].mean()
```

```python
# Quarterly Volume

df.groupby("Quarter")["Volume"].sum()
```

```python
# Visualization
```

```python
# Closing Price Trend

plt.figure(figsize=(14,5))
plt.plot(df["Date"],df["Close"])
plt.xticks(rotation=90)
plt.title("Closing Price Trend")
plt.show()
```

```python
# Trading Volume

plt.figure(figsize=(14,5))
plt.bar(df["Year"],df["Volume"])
plt.title("Trading Volume")
plt.show()
```

```python
# Average Close by Year

df.groupby("Year")["Close"].mean().plot(kind="bar")

plt.show()
```

```python
# Monthly Average Close

df.groupby("Month")["Close"].mean().plot(kind="line")

plt.show()
```

```python
# Distribution of Closing Price

plt.hist(df["Close"],bins=30)

plt.show()
```

```python
# Correlation

sns.heatmap(df.corr(numeric_only=True),annot=True)

plt.show()
```

```python
# Boxplot

plt.scatter(df["Open"],df["Close"])
plt.xlabel("Open")
plt.ylabel("Close")
plt.show()
```

```python
## Business Insights

• Highest stock price was observed during the analysis period, indicating strong market performance on specific trading days.

• Trading volume peaked on certain dates, suggesting increased investor activity and market participation.

• The average closing price fluctuated over the years, reflecting changing market conditions.

• Some months consistently recorded higher average closing prices compared to others.

• Daily return analysis revealed periods of high market volatility.

• A strong positive correlation was observed among the Open, High, Low, and Close prices.
```
